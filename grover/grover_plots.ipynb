```python
import pandas
```

```python
times = pandas.read_csv(open("grover_jsons/times.csv"))
```

```python
import matplotlib.pyplot as plt
```

```python
times
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,9))

ax.set_xlabel("Num of qubits in Grover's search")
ax.set_ylabel("Slicing time", color="orchid")
ax.plot(times["qubit_num"], times["total_time_s"], '--o', ms= 2,lw=1, color="orchid", label="Slicing time")
ax.tick_params(axis='y', labelcolor="orchid")

ax2 = ax.twinx()
ax2.plot(times["qubit_num"], times["slices_count"], '--*', color="slateblue")
ax2.set_ylabel("Slices Count", color="slateblue")

fig.legend(loc="lower left")
# ax2.plot(times["qubit_num"], times["slices_count"], '--.', label="Slices count")
# ax.set_xlabel("Num of qubits in Grover's")
# ax.set_ylabel("Time taken (s)")
# ax2.set_ylabel("Instruction/Slice counts")
plt.legend()
```
